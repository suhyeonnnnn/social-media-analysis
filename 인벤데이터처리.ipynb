```python
import glob
from datetime import date, timedelta

path = "C:/Users/Hannah Kim/Documents/외대/21-2/빅데이터/프로젝트/인벤 데이터/*.xlsx"

for fname in glob.glob(path):
    df=pd.read_excel(fname,header= 0)
    
    name = fname.split('인벤 데이터\\')[1]
    name = name.split('.xlsx')[0]

    df['GameName'] = name
    df_original = df_original.append(df)
```

```python
df = df_original
```

```python
df['date'] = df['date'].str.replace('(','')
```

```python
lst = list(df['date'])
for i in range(len(lst)):
    lst[i] = lst[i].split(' ')[0]
lst
```
