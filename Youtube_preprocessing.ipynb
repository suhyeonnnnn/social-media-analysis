```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("78games.csv")
```

```python
data
```

### Dataframe 형식 전처리

```python
# 필요한 부분만 Slicing
data = data.loc[:, ['VideoDate', 'Comment', 'GameName']]
```

```python
data
```

```python
# Column 이름 변경
data.columns = ['date', 'text', 'GameName']
```

```python
data
```

```python
#Date 행 변수 문자열에서 Datetime 변수로 변환

data['date'] = data['date'].str.replace('[가-힣:]','',regex=True)
data['date'] = pd.to_datetime(data['date'])
```

```python
data
```

### 텍스트 전처리
- 같은 날짜에 반복되는 댓글 제외
- time stamp (ex) 2:56, 14:02) 삭제
- 영어, 특수문자, ㅋㅋㅋ, ㅎㅎㅎ 등의 줄임말 제외
- 글자 수 5자 이하, 100자 이상 댓글 제외

```python
#text와 timestamp 열을 기준으로 중복된 데이터를 삭제
data.drop_duplicates(subset=['date','text'], inplace=True)
```

```python
# 약 700000 개 데이터 제외
data
```

```python
# time stamp (ex) 2:56, 14:02) 삭제
data['text'] = data['text'].str.replace('[0-9]{1,3}:[0-9]{1,3}','',regex=True)
```

```python
# 영어, 특수문자, 자모음 단어 제외

#한글이 아니면 빈 문자열로 바꾸기
data['text'] = data['text'].str.replace('[^0-9가-힣\s]','',regex=True)

#문자열 교체로 인한 공백 제거
data['text'] = data['text'].str.strip()
data['text'] = data['text'].str.replace('[\s+]',' ',regex=True)

#빈 문자열 NAN 값으로 바꾸기
data = data.replace({'': np.nan})
data = data.replace(r'^\s+', None, regex=True)

#NAN 이 있는 행은 삭제
data.dropna(how='any', inplace=True)

#인덱스 정리
data = data.reset_index (drop = True)
```

```python
# 약 800000 개 데이터 제외
data
```

```python
# 글자 수 5자 미만, 100자 이상 댓글 제외

drop_index = data[(data['text'].str.len() < 5) | (data['text'].str.len() > 200)].index
data = data.drop(drop_index)

#인덱스 정리
data = data.reset_index (drop = True)
```

```python
# 숫자만 존재하는 댓글 제외
drop_index = data[data['text'].str.match(r'^[0-9\s]*$')].index
data = data.drop(drop_index)

#인덱스 정리
data = data.reset_index (drop = True)
```

```python
# 약 300000 개 데이터 제외
data
```

```python
data.to_csv("YoutubeComment_preprocessed.csv", index=False)
```
